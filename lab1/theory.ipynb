```python
from PIL import Image
img = Image.open('rrr.jpg')
img.show()
img.save('aaa.png')
```

```python
dim = (100,100,400,400)
crop_img = img.crop(dim)
crop_img.show()
```

```python
grayscale = img.convert('L')
grayscale.show()
```

```python
resize_img = img.resize((200,200))
resize_img.show()
```

```python
rotate_img = img.rotate(90)
rotate_img.show()
```

```python
from PIL import ImageEnhance
enhancer_brightness = ImageEnhance.Brightness(img)
enhancer_brightness.enhance(2).show()
```

```python
enhancer_contrast = ImageEnhance.Contrast(img)
enhancer_contrast.enhance(2).show()
```

```python
img.getpixel((100,100))
```

```python
img.putpixel((100,100),(44,55,66))
img.getpixel((100,100))
```
